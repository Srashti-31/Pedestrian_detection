```python
#import the necessary packages
import numpy as np
import cv2
 
# initialize the HOG descriptor/person detector
hog = cv2.HOGDescriptor()
hog.setSVMDetector(cv2.HOGDescriptor_getDefaultPeopleDetector())
x = []
y=[]
width = []
height = [] 
temp = []

cv2.startWindowThread()

# open webcam video stream
cap = cv2.VideoCapture(0)

# the output will be written to output.avi
out = cv2.VideoWriter('output.avi',cv2.VideoWriter_fourcc(*'MJPG'),15.,(640,480))

while(True):
    # Capture frame-by-frame
    ret, frame = cap.read()

    # resizing for faster detection
    frame = cv2.resize(frame, (640, 480))
    # using a greyscale picture, also for faster detection
    gray = cv2.cvtColor(frame, cv2.COLOR_RGB2GRAY)

    # detect people in the image
    # returns the bounding boxes for the detected objects
    boxes, weights = hog.detectMultiScale(frame, winStride=(8,8) )#finL CROOSSECTION 8X8 

#     boxes = np.array([[x, y, x + w, y + h] for (x, y, w, h) in boxes])#x+h = right se bottom ke coordinate

    for (xA, yA, w, h) in boxes:
        # display the detected boxes in the colour picture
#         print(xA)
#         print(yA)
        cv2.rectangle(frame, (xA, yA), (xA+w, yA+h),(0, 255, 0), 2)
        
        x.append(xA)
        y.append(yA)
        width.append(w)
        height.append(h)
        
        t=[]
        t.append(xA)
        t.append(yA)
        t.append(w)
        t.append(h)
        temp.append(t)
#         print("width",w)
#         print("height",h)

#     print("Before",temp)  

        temp.sort() 
    
#     print("After",temp)
#     print(x)
    
    for i in range(0,len(temp)-2):#-2 === array out of bound
        current = temp[i]
        lastx = current[-2]+current[0]
#         lasty = current[-2]
        
#         print("current -2=",current[-2])

        n = temp[i+1]
        
        startx = n[0]
#         print("startx",n[-2])
#         starty = n[1]


#         print("startx=",startx)
        tester = abs(current[0]-n[0])+abs(current[1]-n[1])+abs(current[2]-n[2])+abs(current[3]-n[3])
        value = 1
        
        if(startx<=lastx and current[-1]>current[-2]>150 and tester >50 
           and abs(current[0]-n[0])>value and abs(current[1]-n[1])>value and 
           abs(current[2]-n[2])>value and abs(current[3]-n[3])>value):
            
            cv2.rectangle(frame, (xA, yA), (xA+w, yA+h),(0,0,255), 2)
            print("Close")
        else:
            print("Far Away")
            
            
    # Write the output video 
    out.write(frame.astype('uint8'))
    # Display the resulting frame
    cv2.imshow('frame',frame)
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

# When everything done, release the capture
cap.release()
# and release the output
out.release()
# finally, close the window
cv2.destroyAllWindows()
cv2.waitKey(1)
```
